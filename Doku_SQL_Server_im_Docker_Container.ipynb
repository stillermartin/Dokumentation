## **Erstellen eines SQL Server Containers (Linux)**

Mit dem Befehl



kann ermittelt werden, welcher Container bereits installiert ist und in welchem Zustand er sich zur Zeit befindet.



Um nun den Container mit dem SQL Server zu erstellen, ist folgender Befehl nötig:



- **docker run**                                                                \[der allgemeine Befehl um einen Container zu erstellen\]
- **\-e**                                                                               \[Hinweis auf Übergabeparameter\]
- **"ACCEPT\_EULA=Y"**                                                  \[Akzeptieren der Enduserlizenz (**E**nd **U**ser **L**icence **A**greement)\]
- **"SA\_PASSWORD="**                                                  \[Administratorenpasswort (Es wird eine SQL Anmeldung erstellt)\]
- **\-p 1402:1433**                                                            \[Angabe über welchen Port für die DB Instanz zu erreichen ist (hier wird Port 1433 nach Port 1402 umgelenkt)\]
- **\--name**                                                                      \[optional: erscheint nur für die Zuordnung bei dem Aufruf **docker ps -a**\]
- **\-d**                                                                               \[das Containerprogramm läuft im Hintergrund\]
- **mcr.microsoftcom/mssql/server:2019-latest.**       \[Name des Images, woraus der Container gebildet werden soll. Hinter dem "**:**" befindet sich das Tag, so dass das Image eindeutig ist\]

Nach dem Start des Befehls, wird der Container auf Basis eines Linuxderivates (Redhat oder Ubuntu) inkl. eines SQL Servers erstellt. Zur Bestätigung erhält man eine TAG Nummer.



Mit dem Befehl **docker ps -a** kann man überprüfen ob der Container wunschgemäß erstellt wurde und ordnungsgemäß läuft.
