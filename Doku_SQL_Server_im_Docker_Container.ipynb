Zunächst muss sichergestellt sein, dass der Docker Dienst (Service) gestartet ist und einwandtfrei läuft.

  

  

  

Ist das nicht der Fall - Docker oder Docker Desktop Service ist installiert - dann den Dienst starten.

  

Sollte Docker/Docker Desktop Service noch nicht installiert sein, muss zunächst die Installation vorgenommen werden.
